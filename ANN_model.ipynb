```python
import tensorflow as tf
print(tf.__version__)
# import some basic libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('Churn_Modelling.csv')
dataset.head()
```

```python
# Divide the dataset into independent and dependent features
X=dataset.iloc[:,3:13]
y=dataset.iloc[:,13]
```

```python
X.head()
```

```python
y.head()
```

```python
# Feature Engineering
geography=pd.get_dummies(X['Geography'],drop_first=True)
gender=pd.get_dummies(X['Gender'],drop_first=True)
```

```python
# concatenate these variable with dataframe
X=X.drop(['Geography', 'Gender'], axis=1)
X.head()
```

```python
pd.concat([X, geography,gender],axis=1)
```

```python
# Splitting the dataset into Training Set and Test Set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
# Part- 2now lets create ANN
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LeakyReLU, PReLU, ELU, LeakyReLU
from tensorflow.keras.layers import Dropout
```

```python
# Lets initiaze the ANN
classifer=Sequential()
```

```python
# Adding the input Layer
classifer.add(Dense(units=11,activation='relu'))
```

```python
# Adding the first hidden Layer
classifer.add(Dense(units=7,activation='relu'))
```

```python
 # Adding the second hidden Layer
classifer.add(Dense(units=6,activation='relu'))
```

```python
# Adding the output layer
classifer.add(Dense(1, activation='sigmoid'))
classifer.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
```

```python
import tensorflow as tf
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0,
    patience=0,
    verbose=0,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0,
)
```

```python
model_history=classifer.fit(X_train, y_train,validation_split=0.33, batch_size=10, epochs=20)
```
